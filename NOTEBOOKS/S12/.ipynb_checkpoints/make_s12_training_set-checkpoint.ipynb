```python
# baker -> place
s11 = {
    'Peter' : 1,
    'Dave' : 2,
    'Laura': 2,
    'Hermine' : 3,
    'Marc' : 4,
    'Lottie' : 5,
    'Mark' : 6,
    'Linda' : 7,
    'Sura' : 8,
    'Rowan' : 9,
    'Mak' : 10,
    'Loriea' : 11
}
```

```python
import pandas as pd
import numpy as np
from scipy import interp
import scipy.stats
from sklearn.preprocessing import QuantileTransformer
from datetime import datetime
```

```python
def timestamp(): return datetime.today().strftime('%Y%m%d')
```

```python
def quantile_scale(df,feats):
    qua = df
    scaler = QuantileTransformer(
        n_quantiles=10,
        random_state=42,
        ignore_implicit_zeros=True, #sparse matrix
    )
    # fit the scaler
    scaler.fit(qua[feats])
    # transform values
    qua[feats] = scaler.transform(qua[feats])
    return qua
```

```python
df = pd.read_csv('../S11/deepbake_features_s11_e10.tsv',sep='\t')
df['place'] = df['baker'].map(s11)
```

```python
tech = pd.read_csv("../../RESULTS/deepbake_features.20201016.tsv",sep='\t')
df2 = pd.concat([tech[df.columns],df])
df2.to_csv('../../RESULTS/deepbake_features.20210926.tsv',sep='\t',index=False)
```

```python
def scale_gbbo(episode):
    df = pd.read_csv(f'../S11/deepbake_features_s11_e{episode}.tsv',sep='\t')
    feats = ['tech_mean','tech','mean_star','star','mean_good','good','mean_bad','bad']
    df2 = df.loc[df['episode']==episode].copy()
    df2 = quantile_scale(df2,feats)
    return df2
```

```python
scale_gbbo(10)
```
